```python
import openpyxl
import pandas as pd
def read_excel(file_name, h):
  dfs = pd.concat(pd.read_excel(file_name, sheet_name=None), ignore_index=True)

  return dfs

dfs = read_excel('/content/gdrive/MyDrive/graph-project/OBR/Final CA125 data.xlsx',2)

with open('/content/gdrive/MyDrive/graph-project/prova.txt', 'w') as f:
    for index, row in dfs.iterrows():
      item=row['Image No.'] + '\\t' + row['Treatment effect'] 
      f.write("%s\n" % item)
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(dfs, test_size=0.2)
train, val= train_test_split(train, test_size=0.1)

bags = [train, val, test]
split = ['train', 'val', 'test']
i=0

for bag in bags:
  with open('/content/gdrive/MyDrive/graph-project/OBR/'+split[i]+'.txt', 'w') as f:
      for index, row in bag.iterrows():
        item=row['Image No.'] + '\\t' + row['Treatment effect'] 
        f.write("%s\n" % item)
  i+=1
```
