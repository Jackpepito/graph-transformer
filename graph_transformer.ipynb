```python
import os
import csv
path_list = []

for path, subdirs, files in os.walk("/content/gdrive/MyDrive/graph-project/OBR_patches"):
    for name in files:
        if not subdirs:
          p = []
          p.append(os.path.join(path, name))
          path_list.append(p)

# opening the csv file in 'a+' mode
file = open("/content/gdrive/MyDrive/graph-project/graph-transformer/feature_extractor/all_patches.csv", 'a+', newline ='')

print(path_list[0])
# writing the data into the file
with file:   
    write = csv.writer(file)
    write.writerows(path_list)

file.close()
```
